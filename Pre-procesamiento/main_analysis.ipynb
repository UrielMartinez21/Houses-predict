## Correlación de Pearson (para variables numéricas)

```python
"""
La correlación de Pearson mide la relación lineal entre dos variables. Como se esta utilizando variables numéricas 
(como número de habitaciones, baños, área), se puede calcular la correlación de Pearson entre cada una de ellas y 
el precio.
"""
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar los datos en un DataFrame
data = pd.read_csv('../Dataset/Mosaico/dataset.csv')

# Eliminar las columnas que no se van a utilizar
data = data.drop(['mosaic_image'], axis=1)

# Calcular la correlación de Pearson entre las variables numéricas
correlation_matrix = data.corr()

# Visualizar la matriz de correlación con seaborn
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title('Matriz de Correlación')
plt.show()

# Mostrar las correlaciones con el precio
print(correlation_matrix['price'].sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt

# Plot the relationship between each tabular feature and the price
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Bedrooms vs Price
axes[0, 0].scatter(data['bedrooms'], data['price'])
axes[0, 0].set_title('Bedrooms vs Price')
axes[0, 0].set_xlabel('Bedrooms')
axes[0, 0].set_ylabel('Price')

# Bathrooms vs Price
axes[0, 1].scatter(data['bathrooms'], data['price'])
axes[0, 1].set_title('Bathrooms vs Price')
axes[0, 1].set_xlabel('Bathrooms')
axes[0, 1].set_ylabel('Price')

# Area vs Price
axes[1, 0].scatter(data['area'], data['price'])
axes[1, 0].set_title('Area vs Price')
axes[1, 0].set_xlabel('Area (sq ft)')
axes[1, 0].set_ylabel('Price')

# Zipcode vs Price
axes[1, 1].scatter(data['zipcode'], data['price'])
axes[1, 1].set_title('Zipcode vs Price')
axes[1, 1].set_xlabel('Zipcode')
axes[1, 1].set_ylabel('Price')

plt.tight_layout()
plt.show()
```

```python
"""
Bedrooms vs Price: Parece haber una ligera tendencia positiva; más habitaciones tienden a 
asociarse con precios más altos, aunque hay mucha variabilidad.

Bathrooms vs Price: Similar a la relación con las habitaciones, más baños suelen corresponder 
a precios más altos, pero también hay dispersión.

Area vs Price: Esta relación muestra una tendencia más clara; casas con mayor área tienden a tener 
precios más altos.

Zipcode vs Price: El código postal parece no mostrar una tendencia clara con el precio, lo que sugiere 
que la ubicación específica dentro de un mismo rango de códigos postales puede tener más importancia.
"""
```
